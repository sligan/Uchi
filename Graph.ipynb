```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'RandomYear': [1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020],
                   'RandomNumber': [4, 6, 3, 8, 9, 6, 2, 8, 5, 10, 7]})


class BaseGraph:
    def __init__(self, dataframe, name=None, x_name=None, y_name=None, color=None):
        self.dataframe = dataframe
        self.name = name
        self.x_name = x_name
        self.y_name = y_name
        self.color = color

    def plot(self, marker=None, linestyle=None, reverse=None):
        if reverse is not None:
            plt.plot(self.dataframe[self.dataframe.columns[1]], self.dataframe[self.dataframe.columns[0]],
                     color=self.color,
                     marker=marker,
                     linestyle=linestyle
                     )
        else:
            plt.plot(self.dataframe[self.dataframe.columns[0]], self.dataframe[self.dataframe.columns[1]],
                     color=self.color,
                     marker=marker,
                     linestyle=linestyle
                     )
        plt.title(self.name)
        plt.xlabel(self.x_name)
        plt.ylabel(self.y_name)
        plt.show()

        
        
class BarGraph(BaseGraph):
    def plot(self, marker=None, linestyle=None, width=5, reverse=None):
        if reverse is not None:
            plt.bar(self.dataframe[self.dataframe.columns[1]], self.dataframe[self.dataframe.columns[0]],
                    color=self.color,
                    width=width
                    )
        else:
            plt.bar(self.dataframe[self.dataframe.columns[0]], self.dataframe[self.dataframe.columns[1]],
                    color=self.color,
                    width=width
                    )
        plt.title(self.name)
        plt.xlabel(self.x_name)
        plt.ylabel(self.y_name)
        plt.show()


```

```python
# задаем общие параметры, которые подходят обеим графикам.
fig = BaseGraph(df, name='GraphName', x_name='any X name', y_name='any Y name', color='green')
# при построении графика можно указать параметры, которые подходят именно этому типу графика.
fig.plot(marker='o', linestyle='--')
```

```python
# при reverse=True оси графика будут перевернуты
fig.plot(linestyle='-.', reverse=True)
```

```python
# график без дополнительных параметров.
fig.plot()
```

```python
# аналогично прошлому графику.
fig_2 = BarGraph(df, name='GraphName', x_name='other X name', y_name='other Y name', color='black')
fig_2.plot(width=1)
```

```python
# график без параметров.
fig_2.plot()
```

```python
# reverse
fig_2.plot(width=1, reverse=True)
```

```python
# графики только с одним обязательным параметром "dataframe".
fig_3 = BaseGraph(df)
fig_3.plot()
```

```python
fig_4= BarGraph(df)
fig_4.plot()
```
